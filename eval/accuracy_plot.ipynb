```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read from CSV
model_accuracy = pd.read_csv("../../../results-nemenyi.csv")

experiment = ['No Resize','Resize 40','Resize 40 \n w/ Margin 1.3']
plt.figure(figsize=(7,5))

best_means = np.array([model_accuracy[column][9:].mean() for column in model_accuracy.columns])
min_for_print = best_means[np.argsort(-best_means)[5]]

for column in model_accuracy.columns:
    plt.plot(experiment,model_accuracy[column][9:],'o-')
#    if model_accuracy[column][9:].mean() > min_for_print: 
#        plt.plot(experiment,model_accuracy[column][9:],'o-')
#    else:
#        plt.plot(experiment,model_accuracy[column][9:],'o-',color='dimgray')
    
plt.ylabel('Accuracy %')  # labels y axis
#plt.legend((model_accuracy.columns))
plt.title('Performance of InceptionResNet-V1 on the ICB-RW')
lgd = plt.legend((model_accuracy.columns), bbox_to_anchor=(1, 0.85))
plt.savefig('ICB-RW_results.pdf', bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
