### Plotting the Losses

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Read from CSV
model_loss = pd.read_csv("../results/SRCNN_coord_x4_train_results.csv")

x = model_loss['Epoch']
y = model_loss['Avg. Loss']

#plt.figure(figsize=(12, 6))
plt.plot(x, y)  # plots the x and y
# rotates the x axis ticks 90 degress and font size 10
#plt.xticks(rotation=90, fontsize=10)
plt.title('SRCNN Coord')  # prints the title on the top
plt.ylabel('MSE Loss')  # labels y axis
plt.xlabel('Epoch')  # labels x axis
plt.savefig('figures/SRCNN_coord_loss.pdf')
```

```python
# Read from CSV
model_loss = pd.read_csv("../results/SRGAN_coord_FaceLossID_x4_train_results.csv")

x = model_loss['Epoch']
y1 = model_loss['Loss_D']
y2 = model_loss['Loss_G']

#plt.figure(figsize=(12, 6))
plt.plot(x, y1)
plt.plot(x, y2)
plt.title('SRGAN Coord with Face Loss')  # prints the title on the top
plt.ylabel('Loss')  # labels y axis
plt.xlabel('Epoch')  # labels x axis
plt.legend(('Discriminator Loss', 'Generator Loss'))
plt.savefig('figures/SRGAN_coord_faceloss.pdf')
```
