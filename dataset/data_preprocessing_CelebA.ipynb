```python
import numpy as np
import matplotlib.pyplot as plt
import os
import random
import PIL
from PIL import Image, ImageDraw

# MTCNN Detection
from facenet_pytorch import MTCNN, extract_face
```

```python
folder = 'test/'

img_path = '/media/angelo/DATEN/Datasets/CelebA/Img/' + folder
img_path_HR = '/media/angelo/DATEN/Datasets/CelebA/HR/' + folder
img_path_LR_56 = '/media/angelo/DATEN/Datasets/CelebA/LR_56/' + folder
img_path_LR_112 = '/media/angelo/DATEN/Datasets/CelebA/LR_112/' + folder
```

```python
if not os.path.exists(img_path_HR):
    os.mkdir(img_path_HR)

if not os.path.exists(img_path_LR_56):
    os.mkdir(img_path_LR_56)

if not os.path.exists(img_path_LR_112):
    os.mkdir(img_path_LR_112)    
```

```python
def load_image(filename):
    # load image from file
    image = Image.open(filename)
    # convert to RGB, if needed
    #image = image.convert('RGB')
    # convert to array
    #pixels = np.asarray(image)
    return image
```

```python
def load_faces(directory):
    face_detector = MTCNN(keep_all=True)
    
    for i, image_path in enumerate(os.listdir(directory)):
        image = load_image(directory + image_path)
        
        face, _ = face_detector.detect(image)
        isEmpty = (str(type(face)) == "<class 'NoneType'>")
        
        if isEmpty: # Checking if faces were detected
            print('Subject ' + image_path + ' did not work.')
        
        if not isEmpty:
            q = random.randint(10,70)
            image = image.crop((face[0][0],face[0][1],face[0][2],face[0][3]))
            image_HR = image.resize((224,224), resample=PIL.Image.BICUBIC)
            image_112 = image.resize((112,112), resample=PIL.Image.BILINEAR)
            image_56 = image.resize((56,56), resample=PIL.Image.BILINEAR)
            
            image_HR.save(img_path_HR + image_path)
            image_112.save(img_path_LR_112 + image_path, quality=q)
            image_56.save(img_path_LR_56 + image_path, quality=q)
        
        if i % 100 == 0:
            print('{}/{} of images processed.'.format(i+1,len(os.listdir(directory))))
        
        
```

```python
load_faces(img_path)
```
